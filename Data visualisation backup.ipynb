```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import boxplot, show   #libraries req for boxplot
%matplotlib inline

df=pd.read_csv("IBRD_ loan_dataset_copy.csv")
df.head()
```

```python
#drop unnecessary columns

to_drop=['Currency of Commitment','Due 3rd Party','Loans Held','Agreement Signing Date','Board Approval Date','Effective Date (Most Recent)','Closed Date (Most Recent)','Last Disbursement Date','Undisbursed Amount','Disbursed Amount','Country Code','Guarantor Country Code','Exchange Adjustment','Borrower\'s Obligation','Due to IBRD','Repaid 3rd Party','Sold 3rd Party']
df.drop(to_drop, inplace=True, axis=1)
```

```python
df
```

```python
# to replace all the 0s in numerical columns with NaNs

df.replace(0,np.nan,inplace=True)
```

```python
df.head()
```

```python
#interpolation

Cancelled_Amount_first = df['Cancelled Amount'].index[4]
Cancelled_Amount_last = df['Cancelled Amount'].index[16]
df.loc[Cancelled_Amount_first:Cancelled_Amount_last, 'Cancelled Amount'] = df.loc[Cancelled_Amount_first:Cancelled_Amount_last, 'Cancelled Amount'].interpolate()
print(df)
```

```python
#fill all the NaNs with the previous values in categorical columns 

df=df.fillna(method='ffill')
df
```

```python
#Replacing all values in numerical columns with mean

#Interest Rate

int_mean=df['Interest Rate'].mean(axis=0)
interest_mean=round(int_mean,3)
print("Mean:",interest_mean)

int_std=df['Interest Rate'].std(axis=0)
interest_std= round(int_std,3)
print("Standard Deviation:",interest_std)

df['Interest Rate'].replace(np.nan,interest_mean,inplace=True)

print(df['Interest Rate'])
```

```python
values=df['Interest Rate']


plt.boxplot(values, whis='range', meanline=True, showmeans=True, vert=False)   #to consider outliers and show the mean of the datapoints
plt.title('Distribution of Interest Rates')
plt.show()
```
